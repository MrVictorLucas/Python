# Automação de processos com Python

## Problema a ser resolvido - passo a passo

## Passo 1 - Buscar automaticamente os dados das ações

- Buscar automaticamente os dados das ações
- Gerar as análises de forma automática
- Enviar um e-mail para o nosso gestor

```python
import yfinance
import matplotlib
```

```python
codigo = input("Digite o código da ação desejada: ")
dados = yfinance.Ticker(codigo).history("24mo")
fechamento = dados.Close
fechamento.plot()
```

## Passo 2 - Gerar as análises de forma automática

- Análise dos últimos 24 meses
- Cotação máxima
- Cotação mínima
- Cotação atual

```python
maxima = fechamento.max()
minima = fechamento.min()
atual = fechamento[-1]
```

## Passo 3 - Enviar um e-mail para o nosso gestor

- Abrir uma nova aba no navegador (CRTL + T)
- Digitar o endereço do E-mail e dar Enter
- Clicar no botão escrever
- Colocar o E-mail do destinatário (TAB)
- Colocar o assunto (TAB)
- Escrever o corpo do e-mail
- Clicar em enviar

```python
import pyautogui as pa
import pyperclip as pp
```

```python
import time
time.sleep(2)

#pausa entre cada command
pa.PAUSE = 3

#abrir o navegador e criar uma janela
pa.click(x=506, y=14)
pa.hotkey("ctrl", "t")

#digitar o endereço de email e dar enter
pp.copy("https://mail.google.com/mail/u/1/#inbox")
pa.hotkey("ctrl", "v")
pa.hotkey("enter")

#clicar no botão escrever
pa.click(x=138, y=236)

#colocar o email do destinatário (tab)
pp.copy("vgoncalves14vl@gmail.com")
pa.hotkey("ctrl", "v")
pa.hotkey("tab")

#colocar o assunto (tab)
pp.copy("Análises Diárias")
pa.hotkey("ctrl", "v")
pa.hotkey("tab")

#escrever a mensagem
mensagem = f"""
Prezado gestor,

Seguem as análises diárias dos últimos 24 meses da ação {codigo}:

Cotação Máxima: US$ {round(maxima, 2)}
Cotação Mínima: US$ {round(minima, 2)}
Cotação Atual: US$ {round(atual, 2)}

Qualquer dúvida, fico à disposição

"""

pp.copy(mensagem)
pa.hotkey("ctrl", "v")

#clicar pra enviar
pa.click(x=1304, y=1043)
```
